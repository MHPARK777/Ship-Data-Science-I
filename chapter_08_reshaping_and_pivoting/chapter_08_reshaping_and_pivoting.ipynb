```python
file_path = '/content/drive/MyDrive/chapter_08_reshaping_and_pivoting/video_game_sales.csv'
```

```python
video_game_sales = pd.read_csv(file_path)
video_game_sales.head()
```

```python
video_game_sales.melt(id_vars='Name', value_vars='NA').head()
```

```python
regional_sales_columns = ['NA', 'EU', 'JP', 'Other']

video_game_sales.melt(id_vars='Name',
                     value_vars=regional_sales_columns)
```

```python
video_game_sales_by_region = video_game_sales.melt(id_vars='Name',
                                                  value_vars=regional_sales_columns,
                                                  var_name='Region',
                                                  value_name='Sales')

video_game_sales_by_region
```

```python
video_game_sales_by_region = video_game_sales.melt(id_vars='Name',
                                                  value_vars=regional_sales_columns,
                                                  var_name='Region',
                                                  value_name='Sales')

video_game_sales_by_region.head()
```

```python
video_game_sales_by_region.pivot_table(index='Name',
                                      values='Sales',
                                      aggfunc='sum').head()
```

```python
video_game_sales_by_region.pivot_table(index='Name',
                                      values='Sales',
                                      aggfunc='sum').tail()
```

```python
video_game_sales_by_region.pivot_table(index='Name',
                                      values='Sales',
                                      aggfunc='sum')
```

```python
video_game_sales_by_region.pivot_table(index='Name',
                                      values='Sales',
                                      aggfunc='sum').sort_values(by='Sales')
```
